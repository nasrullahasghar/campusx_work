## **Builtin Modules**

**What are Modules?**
- Code libraries in Python.
- Contain functions for use.
  
*Examples:*

- Math: Math ops
- Random: Random numbers
- os: OS interaction
- time: Time tasks

```python
# Math Module
import math

math.factorial(5)
```

```python
math.pi
```

```python
math.e
```

```python
math.floor(3.43)
```

```python
math.ceil(7.3)
```

```python
# Random Module
import random

random.randint(1,100)
```

```python
a = [1, 2, 3, 4, 5]
random.shuffle(a)
print(a)
```

```python
# Time Module
import time

time.time()
```

```python
time.asctime()
```

```python
time.ctime()
```

```python
print("Hello")
time.sleep(2)
print("World")
```

```python
# Datetime Module
import datetime

print(datetime.datetime.now())
```

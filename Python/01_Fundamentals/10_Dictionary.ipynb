# **Dictionary Rules**
1. Dictionary Has no indexing
2. Dictionary is a mutable data type just like sets and list
3. `Keys` should always be immutable and `values` can be mutable and immutable
4. `Keys` should be unique

### **Q2️⃣: Add & Update**

- Start with:

    `car = {"brand": "Toyota", "model": "Corolla", "year": 2018}`


- Add a new key "color" with value "Black".

- Update the "year" to 2022.

- Print the updated dictionary.

```python
car = {"brand": "Toyota", "model": "Corolla", "year": 2018}
car.update({'color':'Black'})
car['year'] = 2022
car
```

### **Q4️⃣: Loop Through Dictionary**

- Given:

    `fruits = {"apple": 100, "banana": 60, "cherry": 150}`

- Print all keys and their values in the format:\
    `👉 apple costs 100`

```python
fruits = {"apple": 100, "banana": 60, "cherry": 150}

for i in fruits:
    print(i,fruits[i])
```

### **Q5️⃣: Using .get() Safely**

- Create a dictionary:

    `marks = {"Math": 88, "Science": 92, "English": 81}`


- Ask the user to enter a subject name and print the marks using .get().
- If the subject doesn’t exist, show:\
    `👉 "Subject not found!"`

```python
marks = {"Math": 88, "Science": 92, "English": 81}
key = input("Enter the subject Name").capitalize()

if key in marks.keys():
    print(marks.get(key))
else:
    print("Subject Not Found")
```

#### **Q7️⃣: Merge Dictionaries**

- Given:

    `dict1 = {"a": 1, "b": 2}`\
    `dict2 = {"c": 3, "d": 4}`


- Merge both dictionaries into a new one and print it.

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

merged_dict = dict1.copy()  
merged_dict.update(dict2)

print(merged_dict)
```

### **Q8️⃣: Count Word Frequency**

- Ask the user to enter a sentence and count how many times each word appears.
- Example:\
    `Input: "I love Python Python"`\
    `Output: {'I': 1, 'love': 1, 'Python': 2}`

| Word             | Before      | `.get(word, 0)` | New Value | Dictionary After                 |
| ---------------- | ----------- | --------------- | --------- | -------------------------------- |
| "I"              | not present | 0               | 1         | {'I': 1}                         |
| "love"           | not present | 0               | 1         | {'I': 1, 'love': 1}              |
| "Python"         | not present | 0               | 1         | {'I': 1, 'love': 1, 'Python': 1} |
| "Python" (again) | present (1) | 1               | 2         | {'I': 1, 'love': 1, 'Python': 2} |


```python
sentence = input("Enter the Sentence").capitalize()
words = sentence.split()
words_count = {}

for word in words:
    words_count[word] = words_count.get(word,0) +1

print(words_count)
```

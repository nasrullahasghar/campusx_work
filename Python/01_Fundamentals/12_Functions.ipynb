# **1. Components of Function**

```python
def is_even(num): # def:"Keyword",is_even:"name" and num:"argument"
    
# Optional docstring
    '''     
    Checks if a number is even or odd.

    Args:
        num: The number to check.
    Returns:
        A string indicating whether the number is "Even", "Odd", 
        or an error message if the input is not a valid number.

    Created by : Mirza
    Last Edited : 24 Oct 2025'''

    # body
    if type(num) == int:
        if num % 2 == 0:
            return "Even"
        else:
            return "Odd"
    else:
        return "Please enter a valid number"
```

```python
X = is_even(4)
print(X)
```

```python
for i in range(1,11):
    print(f'{i} = {is_even(i)}')
```

## **1. Default Argument**

```python
def power(a,b): # The arguments are by default
    return a**b
```

```python
power(2,3)
```

```python
power(2) # we did not want out like this 
```

## **2. Positional Argument**

```python
def power(a=1,b=1): # We specify the position of argument
    return a**b
```

```python
power(1)
```

```python
power(2,3)
```

## **3. Keyword Argument**

```python
power(b=2,a=3) # when we knows about the anly few arguments of function we directly use keywords of arguments
```

```python
def flexi(*number):
    product = 1
    print(number)
    for i in number:
        product = product * i
    print("The Product of the Numbers = ",product)
        
```

```python
flexi(1,2,3)
```

Write a function called greet() that asks for the user’s name and prints:
- 👉 "Hello, [name]! Welcome to Python."
- Then call the function to see the output.

```python
def greet(name):
    if type(name) == str:
        return f"Hello, {name}! Welcome to Python."
    else:
        return "Please Enter a String"
```

Write a function find_max(a, b, c) that takes three numbers and returns the maximum number.

```python
def find_max(a, b, c):
    if a > b and c:
        return a
    elif b > a and c:
        return c
    else:
        return b
```

**Q5️⃣: Factorial Finder**

Write a function factorial(n) that returns the factorial of a number using a for loop.
Example:
👉 factorial(5) → 120

```python
def factorial(n):
    if type(n) == int:
        if n < 0:
            return "Factorial is not defined for negative numbers."
        else:
            result = 1
            for i in range(1,n+1):
                result = result*i
            return f'The Factorial of {n} = {result}'

    else:
        return "Please enter an Integer"
```

```python
factorial(5)
```

**Q6️⃣: Count Vowels in a String**

Write a function count_vowels(text) that returns the number of vowels in the string entered by the user.

```python
text = input("Enter the Sentence").lower()

def count_vowels(text):
    vowels = 'aeiou'
    vowels_count = 0
    for i in text:
        if i in vowels:
            vowels_count+= 1
    return vowels_count

result = count_vowels(text)
print("The Number of vowels = ",result)
```

### Q6. Functions

- Given numbers = [12, 7, 9, 21, 5]

- Print the length of the list.

- Print the maximum and minimum number.

- Print the sum of all numbers.

- Sort the list in ascending order.

```python
list = [12, 7, 9, 21, 5]
print(f"The String has {len(list)} numbers")
print(f"The maximum number is: {max(list)} , and the minimum number is: {min(list)}")
print(f"The sum of all numbers in string: {sum(list)}")
list.sort()
print("Sorted List:",list)
```

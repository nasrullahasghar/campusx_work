# **Sets Rules**
1. Sets do no allow duplicates.
2. Sets have no indexing / slicing
3. Sets do no allow mutable data types
4. Set itself is a mutable data type(means there is no 2D set)

### **Q3. Set Operations**

Let\
A = {1, 2, 3, 4, 5}\
B = {4, 5, 6, 7, 8}

`Find and print:`

- Union of A and B

- Intersection of A and B

- Difference (A - B)

- Symmetric Difference

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("Union of A and B:", A.union(B))
print("Intersection of A and B:", A.intersection(B))
print("Difference (A - B):", A.difference(B))
print("Symmetric Difference:", A.symmetric_difference(B))
```

### **Q6. Challenge**

- Write a program that takes two sets from the user and checks whether one set is a subset or superset of the other.

```python
set1 = set(input("Enter elements for Set 1: "))
set2 = set(input("Enter elements for Set 2: "))

if set1.issubset(set2):
    print("Set1 is a subset of Set2")
elif set1.issuperset(set2):
    print("Set1 is a superset of Set2")
elif set2.issubset(set1):
    print("Set2 is a subset of Set1")
elif set2.issuperset(set1):
    print("Set2 is a superset of Set1")
else:
    print("No subset or superset relationship found")
```

```python
import numpy as np
jackpot = np.random.randint(1,100)

guess = int(input("Enter the Number"))
counter = 1

while guess != jackpot :
    if guess < jackpot:
        print("Guess Higher Number")
    else:
        print("Guess Lower Number")
    guess = int(input("Enter Again Number"))
    counter +=1

print("Congratulations You Guess the Number!")
print(f"You took {counter} iterations")
```

**🔹 Q1. Sum of First N Numbers (Using a while loop)**\
Ask the user to enter a number n. Print the sum of first n natural numbers (i.e. 1+2+3+...+n) using a while loop.

```python
number = int(input("Enter the number"))
i=1
total_sum = 0
while i <= number:
    total_sum += i
    i += 1
print(f"Total Sum from 1 to {number} is equal to {total_sum}")
```

**🔹 Q2. Multiplication Table (Using a for loop)**\
Ask the user to enter a number. Print its multiplication table up to 10 using a for loop.

```python
number = int(input("Enter the number"))
for i in range(1,11):
    print(f"{i} x {number} = {i*number}")
```

**🔄 Updated Q4: Secret Number Game with Hints**

*✅ Requirements:*
- Secret number is 13 (you can also randomize it later).

- Keep asking the user for a number.

- If the guess is correct, print "Secret number found!" and break.

- If the guess is too high, print "Lower number, try again...".

- If the guess is too low, print "Higher number, try again...".

```python
guess = int(input("Enter The Number"))
secret = 13
counter = 1
while guess != secret:
    if guess < secret:
        print("Higher number, try again...")
    else:
        print("Lower number, try again...")
    guess = int(input("Enter The Number Again"))
    counter += 1
print("Secret number found!")
print(f"You took {counter} attempts.")
```

**🔹 Q5. Pattern Printing – Nested Loops**\
Print the following pattern using nested for loops:

```python
*
* *
* * *
* * * *
* * * * *
```

```python
rows = 5
i = 1
for i in range(1,rows+1):
    for j in range(0,i):
        print("*",end=" ")
    print("")
```

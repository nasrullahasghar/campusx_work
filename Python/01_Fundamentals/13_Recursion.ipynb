## **Recursion**
Recursion is a programming technique in which a function calls itself in order to solve a problem. It is often used to break down complex problems into simpler, more manageable subproblems. A recursive function typically has two main components:

`Base Case:` This is the condition under which the recursive calls stop. It prevents the function from calling itself indefinitely.

`Recursive Case:` This is where the function calls itself with a modified argument, moving it closer to the base case.

```python
# Without Recursion
def multiply(a,b):

    result = 0
    for i in range(b):
        result = a + result
    print(result)

multiply(3,4)
```

```python
# With Recursion
def mul(a,b):
    if b==1:
        return a
    else:
        return a + mul(a,b-1)
    
mul(3,4)
```

```python
def factorial(n):

    if (n==0 or n==1):
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
```

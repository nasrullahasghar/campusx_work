# **Tuple Practice Questions**

**Q1. Create & Access**

- Create a tuple of 5 colors.

- Print the first color and the last color using indexing.

```python
colors = ("red","green","blue","yello","black")
print(colors[0])
print(colors[-1])
print("Togather :",colors[0::4])
```

**Q2. Slicing**

- Given numbers = (10, 20, 30, 40, 50, 60)

- Print the tuple from index 2 to 4.

```python
numbers = (10, 20, 30, 40, 50, 60)
numbers[2:5]
```

**Q3. Unpacking**

- Create a tuple with 3 values (name, age, city).

- Unpack them into separate variables and print each one.

```python
unpack = ("name", "age", "city")
(name, age, city) = unpack
print(name)
print(age)
print(city)
```

```python
person = ("Adeel", 22, "Lahore")
(name, age, city) = person
print(f"Name:{name}, Age:{age}, City:{city}")
```

**Q4. Nested Tuple**

- Create a tuple that contains two tuples:
`(("apple", "banana"), (1, 2, 3))`

- Access "banana" and 3 using indexing.

```python
fruits = (("apple", "banana"), (1, 2, 3))
print(fruits[0][1],fruits[1][2])
```

**Q5. Functions**

- Given marks = (55, 72, 91, 66, 72, 88)

- Find the maximum, minimum, and count how many times 72 appears.

```python
marks = (55, 72, 91, 66, 72, 88)
print("Maximum Marks =",max(marks))
print("Minimum Marks =",min(marks))
print("Count of 72 =",marks.count(72))
```

**Q6. Challenge**

Write a Python program to check if a given element exists in a tuple.

```python
numbers = (10, 20, 30, 40, 50, 60)
user_num = int(input("Enter the number to check"))
if user_num in numbers:
    print("Yes the element is Existed")
else:
    print("No the element is Not Existed")
```

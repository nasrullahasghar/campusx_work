**🧠 Q1. Even or Odd Checker**
- Write a program that asks the user to input a number and checks whether it is even or odd.

```python
num = int(input("Enter a number"))

if num % 2 == 0:
    print("Number is Even")
else:
    print("Number is Odd")
```

**🎂 Q2. Age Category Checker**

- Ask the user for their age.

- If age < 13 → print “Child”

- If age is between 13 and 19 → print “Teenager”

- If age is between 20 and 59 → print “Adult”

- If age is 60 or more → print “Senior Citizen”

```python
age = int(input("Enter Your Age"))

if age < 13:
    print("You are a Child")
elif age >= 13 and age <= 19:
    print("You are a Teenager")
elif age > 19 and age <=59:
    print("You are a Adult")

else:
    print("You are a Senior Citizen")
```

**🏆 Q3. Number Comparison Game**
- Ask the user to input two numbers, and print which number is greater or if they are equal.

```python
num1 = int(input("Enter the first Number"))
num2 = int(input("Enter the Second Number"))


if num1 == num2:
    print("Both Numbers are Equal")
elif num1 > num2:
    print("First Number is Greater than Second Number")
else:
    print("Second Number is Greater than First Number")
```

**📚 Q4. Grade Calculator**

Ask the user to input marks (0-100).

Print the grade as per the following:

- 90-100 → A

- 80-89 → B

- 70-79 → C

- 60-69 → D

- Less than 60 → F

- Anything else → “Invalid Marks”

```python
marks = int(input("Enter Your Marks"))

if marks <= 100:

    if marks < 60:
        print("Grade F")

    elif marks >= 60 and marks <=69:
        print("Grade D")

    elif marks >=70 and marks <=79:
        print("Grade C")
    
    elif marks >=80 and marks <=89:
        print("Grade B")

    elif marks >=90 and marks <=100:
        print("Grade A")

else:
    print("Invalid Marks")
```

**📅 Q5. Leap Year Checker**

Write a program to check whether a given year is a leap year or not.\
A leap year is:

- Divisible by 4 ✅

- But not divisible by 100 ❌

- Unless it’s also divisible by 400 ✅

```python
year = int(input("Enter Year"))

if (year % 4 == 0 and year % 100 !=100) or (year % 400 == 0):
    print("It's a Leap Year")

else:
    print("It's not a Leap Year")
```

# **2. Aliasing**

`In Python, aliasing occurs when multiple variables refer to the same object in memory. This means that instead of creating separate copies of an object, you are creating additional names (aliases) that point to the exact same underlying data.`

```python
a = 5
b = a
print("Address of a : ",id(a))
print("Address of b : ",id(b))
```

```python
c = b
id(c)
```

```python
# we can delete the pointer's names
del a
a
```

```python
del b
b
```

```python
# C is not deleted so the address of 5 is stored in c
id(c)
```

```python
a = 6
b = a
b
```

```python
# Here a is pointing a new address
a = 7
a
```

# **6. Mutibility**

| Type         | Mutable/Immutable | Description                                      |
|--------------|-------------------|--------------------------------------------------|
| `int`        | Immutable         | Integer values                                   |
| `float`      | Immutable         | Floating-point numbers                           |
| `str`        | Immutable         | String values                                    |
| `tuple`      | Immutable         | Ordered collection of items                      |
| `list`       | Mutable           | Ordered and changeable collection of items       |
| `dict`       | Mutable           | Collection of key-value pairs                    |
| `set`        | Mutable           | Unordered collection of unique items             |
| `frozenset`  | Immutable         | Immutable version of a set                       |

```python
a = "Hello"
id(a)
```

```python
a = a + "world"
id(a)
```

```python
L = [1,2,3]
id(L)
```

```python
L.append(4)
print(L)
print(id(L))
```

```python
""" In Python, immutable data types (like int, str, and tuple) cannot be changed once created; 
modifying them creates a new object in memory, resulting in a different address.
Mutable data types (like list, dict, and set) can be modified in place; 
changes affect the original object without creating a new memory address."""
```

# **7. Side Effects of Mutibility**

```python
a = [1,2,3]
b = a
print(id(a))
print(id(b))
```

```python
b.append(4)
print(b)
```

```python
# Why it is happening?
print(a)
print(id(a))
print(id(b))
```

```python
'''This is because of mutable nature of list, due to aliasing the both will be changed be 
because of their same address'''
```

```python
# To over come this problem
a = [1,2,3]
b = a[:]
print(id(a))
print(id(b))
```

```python
'''b = a[:] creates a new list b that is a shallow copy of a. 
This means that b contains the same elements as a, but they are stored in a different memory location.'''
```

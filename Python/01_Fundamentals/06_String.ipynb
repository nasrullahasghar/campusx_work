## **Example Programs**

🧠 Q1. Palindrome Checker

Ask the user for a string and check if it is a palindrome (same forward and backward).
Example:

- "madam" → Palindrome ✅

- "python" → Not Palindrome ❌

```python
a = str(input("Enter a Word").lower())
reverse_string = a[::-1]
if a == reverse_string:
    print("The Word is Palindrome")
else:
    print("The Word is Not Palindrome")
```

🧠 Q2. Count Vowels and Consonants

Take a string input from the user and count:

- Total vowels (a, e, i, o, u)

- Total consonants

```python
b = str(input("Enter a Word")).lower()
vowels = "aeiou"

vowels_count = 0
consonants_count  = 0

for i in b:
    if i in vowels:
        vowels_count +=1
    else:
        consonants_count +=1

print(f"There are {vowels_count} vowels in you word") 
print(f"There are {consonants_count} consonants in you word") 
```

🧠 Q4. Find and Replace

Take a string input, ask the user which word to replace, and with what. Then print the new string.

(Hint: use .replace())

```python
c = input("Enter a Sentence").strip()
user_replacement = input("Enter the Word you want to Replace: ").strip()
with_what = input("Which Word you want to Replace with?: ").strip()

if user_replacement in c:
    c = c.replace(user_replacement, with_what)
    print("Updated Sentence: ",c)
else:
    print("Your Word is not in the Sentence")
```

🧠 Q5. Longest Word Finder

Ask the user for a sentence. Find and print the longest word in the sentence.

Example:
Input: "Python is powerful and amazing"

Output: "powerful"

```python
d = input("Enter a Sentence").strip()
words_list = d.split()

longest_word = ''
max_length = 0
for word in words_list:
    if len(word) > max_length:
        longest_word = word 
        max_length = len(word)
print(f"The Longest Word is {longest_word} with length of {max_length}") 
```

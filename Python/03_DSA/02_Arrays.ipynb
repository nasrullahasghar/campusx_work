# **Types of Arrays**
1. **Normal Array**
2. **Referential Array**
3. **Dynamic Array**

## **1. Normal Array**
1.  Can store only one Data Type(Homogenous)
2.  Has Fixed Size(Static Array)

```python
import array as arr

# Get the size of the array from user input
size = int(input("Enter the size of the array: "))

# Create an array of integers
array = arr.array('i', [])

# Loop to append elements to the array
for i in range(size):
    element = int(input(f"Enter element {i + 1} of the array: ")) 
    array.append(element)

# Display the created array
print("The created array is:", array)
```

## **3. Dynamic Array**
- It is used to solve the 2nd problem of Normal Array.

## **Example**
We will make our own python list

```python
import ctypes

class MeraList:
    # Constructor
    def __init__(self):
        self.size = 1
        self.n = 0
        self.A = self.__make_array(self.size)

    # Length 
    def __len__(self):
        return self.n
    
    # Append
    def append(self, item):
        if self.n == self.size:  # Array is full -> resize
            self.__resize(self.size * 2)

        self.A[self.n] = item
        self.n += 1

    # Pop in List
    def pop(self):
        if self.n == 0:
            return 'List is Empty'
        item = self.A[self.n - 1]
        self.n -= 1
        return item
            
    # Clear 
    def clear(self):
        self.n = 0
        self.size = 1
        self.A = self.__make_array(self.size)  # Reset array

    # Find
    def find(self, item):
        for i in range(self.n):
            if self.A[i] == item:
                return i
        return None  # Changed to return None if not found

    # Print
    def __str__(self):
        return '[' + ', '.join(str(self.A[i]) for i in range(self.n)) + ']'
    
    # Indexing in list
    def __getitem__(self, index):
        if 0 <= index < self.n:  
            return self.A[index]
        else:
            raise IndexError("Index out of bounds")
        
    # Array Insertion
    def insert(self, pos, item):
        if self.n == self.size:  # Resize if needed
            self.__resize(self.size * 2)

        for i in range(self.n, pos, -1):
            self.A[i] = self.A[i - 1]  # Shift elements right

        self.A[pos] = item  # Insert the new item
        self.n += 1  # Increment the size of the list

    # Array Deletion
    def __delitem__(self, pos):
        if pos < 0 or pos >= self.n:
            raise IndexError("Index out of bounds")

        for i in range(pos, self.n - 1):
            self.A[i] = self.A[i + 1]  # Shift elements left

        self.A[self.n - 1] = None  # Clear the last element
        self.n -= 1  # Decrement the size

    # Resize Method for Append 
    def __resize(self, new_capacity):
        B = self.__make_array(new_capacity)
        for i in range(self.n):
            B[i] = self.A[i]
        self.A = B
        self.size = new_capacity

    # Make 
    def __make_array(self, capacity):
        return (capacity * ctypes.py_object)()
```

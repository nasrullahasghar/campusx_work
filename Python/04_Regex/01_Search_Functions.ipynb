## **1.1 Search()**
The `re.search()` function in Python's `re` module is used to search for a pattern in a string. Here are its main properties:

### **Syntax**
re.search(pattern, string, flags=0)

**Parameters:**

- `pattern:` A string or a compiled regular expression that defines the search pattern.
- `string:` The string where the search will take place.
- `flags:` Optional. A bitwise OR (|) of flags that specify additional options (e.g., re.IGNORECASE).

```python
import re
pattern = r"Python" #We will use r string to ignore escape sequences
data = "Python is very powerful and it has many features"
finded = re.search(pattern,data)


print(finded)
print(type(finded))

if finded:
    print("Found")
else:
    print("Not Found")
```

## **1.2 match()**
The `re.match()` function in Python's `re` module is used to determine if a regular expression pattern matches at the beginning of a string. Here are its main properties:

### **Syntax**

re.match(pattern, string, flags=0)

**Parameters**
- `pattern:` A string or a compiled regular expression that defines the search pattern.
- `string:` The string to be searched.
- `flags:` Optional. A bitwise OR (|) of flags that specify additional options (e.g., re.IGNORECASE).

```python
import re
pattern = r"PYTHOn"
data = "Python is very powerful and it has many features"
finded = re.match(pattern,data,re.IGNORECASE) # re.IGNORECASE used to make case insensitive

print(finded)

if finded:
    print("Found")
else:
    print("Not Found")
```

## **1.3 finditer()**
The `re.finditer()` function in Python's `re` module is used to find all occurrences of a pattern in a string as an `iterator` of match objects.
- Memory Efficient due to iter as return
### **Syntax**
re.finditer(pattern, string, flags=0)

**Parameters**

- `pattern`: A string or a compiled regular expression that defines the search pattern.
- `string`: The string to be searched.
- `flags`: Optional. A bitwise OR (|) of flags that specify additional options (e.g., re.IGNORECASE).

```python
pattern = re.compile('ab',re.IGNORECASE) # We can use flag only here if we are using compile to make insensitive
data = 'ababaabaababab'
iter = re.finditer(pattern,data)
print(iter)

for find in iter:
    # print(find.start())
    # print(find.group())
    # print(find.end())
    print(f"This is for separate =========>start:{find.start()}, end:{find.end()}, group:{find.group()}")
    print("=============================================================")
    # In this every thing is avaiable
    print("This is for all",find)
```

## **1.4 findall()**
The `re.findall()` function in Python's re module returns all non-overlapping matches of a pattern in a `string` as a `list of strings`.
- Since it returns list so it's not memory efficient
### **Syntax**
re.findall(pattern, string, flags=0)

**Parameters**

- `pattern`: A string or a compiled regular expression that defines the search pattern.
- `string`: The string to be searched.
- `flags`: Optional. A bitwise OR (|) of flags that specify additional options (e.g., re.IGNORECASE).

```python
pattern = re.compile('ab',re.IGNORECASE) # We can use flag only here if we are using compile to make insensitive
data = 'ababaabaababab'
match_list = re.findall(pattern,data)
print(match_list)
```

## **1.5 split()**
The `re.split()` function in Python's re module splits a string by the occurrences of a pattern.

### **Syntax**
re.split(pattern, string, maxsplit=0, flags=0)
**Parameters**

- `pattern:` A string or a compiled regular expression that defines the delimiter pattern.
- `string:` The string to be split.
- `maxsplit:` Optional. Maximum number of splits to perform. If omitted or 0, all occurrences are split.
- `flags:` Optional. A bitwise OR (|) of flags that specify additional options (e.g., re.IGNORECASE).

```python
text = "Hello, world! This is an example. Let's split this string."

# Pattern to split on punctuation and spaces
pattern = r"[,.! ]+"

# Using re.split() to split the string
result = re.split(pattern, text)

# Filtering out any empty strings from the result
result = list(filter(None, result))

# Outputting the result
print(result)
```

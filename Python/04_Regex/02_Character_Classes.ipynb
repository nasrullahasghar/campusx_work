# **Character Classes in Regex**

| Character Class | Description                                                | Equivalent To       |
|------------------|------------------------------------------------------------|---------------------|
| **`\d`**         | Matches any digit character.                               | `[0-9]`             |
| **`\D`**         | Matches any non-digit character.                           | `[^0-9]`            |
| **`\w`**         | Matches any word character (alphanumeric and underscore). | `[a-zA-Z0-9_]`      |
| **`\W`**         | Matches any non-word character (special characters).      | `[^a-zA-Z0-9_]`     |
| **`\s`**         | Matches any whitespace character.                          | `[ \t\n\r\f\v]`     |
| **`\S`**         | Matches any non-whitespace character.                     | `[^ \t\n\r\f\v]`    |
| **`.`**          | Matches any single character except a newline.            | (unless re.S flag is set) |

| Character Class    | Description                                           |
|---------------------|-------------------------------------------------------|
| **`[abc]`**        | Matches any one of the characters 'a', 'b', or 'c'. |
| **`[0-9]`**        | Matches any character that is a digit from '0' to '9'. |
| **`[A-Z]`**        | Matches any uppercase letter from 'A' to 'Z'.       |
| **`[a-z]`**        | Matches any lowercase letter from 'a' to 'z'.       |
| **`[A-Za-z]`**     | Matches any uppercase or lowercase letter.           |
| **`[0-9A-Fa-f]`**  | Matches a hexadecimal digit (0-9, A-F, or a-f).     |
| **`[a-zA-Z0-9]`**  | Matches any alphanumeric character.                  |
| **`[^0-9]`**       | Matches any character that is not a digit.           |
| **`[^A-Za-z]`**    | Matches any character that is not an uppercase or lowercase letter. |

```python
import re
```

```python
pattern = r'[a-z0-9]'
data = "This Mobile costs me about $100.0"
iter = re.finditer(pattern,data,re.IGNORECASE)

match_list = re.findall(pattern,data,re.IGNORECASE)
print("This is match list using findall:",match_list)
for find in iter:
    print(find)
```

```python
pattern = r'[^a-z0-9 ]' # ^ use to exclude
data = "This Mobile costs me about $100.0"
iter = re.finditer(pattern,data,)

match_list = re.findall(pattern,data,)
print("This is match list using findall:",match_list)
for find in iter:
    print(find)
```

```python
# Condition for Password
# Mix uppercase (), lowercase (), numbers ()

password = input("Enter the Password")

pattern = r"[0-9A-Z]"

data = re.findall(pattern,password)

if data:
    print("Valid Password")
else:
    print("Invalid Password")
```

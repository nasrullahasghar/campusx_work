### **5. Write a program that will reverse a four digit number.Also it checks whether the reverse is true.**

```python
num = input("Enter a number: ")

rev_num = num[::-1]

# Display the reversed number
print("Reversed number:", int(rev_num))

if num == rev_num:
    print(True)
else:
    print(False)
```

### **6. Write a program that will tell whether the number entered by the user is odd or even.**

```python
def even_odd(num):
    if num % 2 == 0:
        print(f"Your Number {num} is Even.")
    else:
        print(f"Your Number {num} is Odd.")

even_odd(4)
```

### **7. Write a program that will tell whether the given year is a leap year or not.**

```python
def leap_year(year):
    if (year % 4 == 0 and  year % 100 != 0) or year % 400 == 0:
        print("It's a Leap Year")
    else:
        print("it's not a leap Year")

leap_year(2028)
```

### **8. Write a program to find the euclidean distance between two coordinates.**

```python
# Input the coordinates of the two points
X1 = float(input("Enter X1 of x coordinate: "))
X2 = float(input("Enter X2 of x coordinate: "))
Y1 = float(input("Enter Y1 of y coordinate: "))
Y2 = float(input("Enter Y2 of y coordinate: "))

# Calculate the distance using the distance formula
d = ((X2 - X1)**2 + (Y2 - Y1)**2)**0.5

# Print the result
print("The distance between the points is:", d)
```

## **9. Write a program that take a user input of three angles and will find out whether it can form a triangle or not.**

```python
# User Input
alpha = int(input("Enter Engle Greater than 0"))
beta = int(input("Enter Engle Greater than 0"))
gamma = int(input("Enter Engle Greater than 0"))

# Logic
if alpha<=0 or beta <=0 or gamma<=0:
    print("All angles must be greater than 0")
else:
    if alpha+beta+gamma == 180:
        print("The angles can form a triangle.")
    else:
        print("The angles cannot form a triangle.")
```

```python
def triangle(alpha,beta,gamma):

    if alpha<=0 or beta <=0 or gamma<=0:
        print("All angles must be greater than 0")
    else:
        if alpha+beta+gamma == 180:
            print("The angles can form a triangle.")
        else:
            print("The angles cannot form a triangle.")

triangle(60,80,40)
```

## **10. Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit**

```python
cost = int(input("Enter the Cost Price"))
sell = int(input("Enter the Selling Price"))

profit = sell - cost
loss =  cost - sell

if sell > cost:
    print(f"You are in Profit of {profit} rupees.")

elif sell == cost:
    print("You are neither in Profit nor in Loss")
else:
    print(f"You are in loss of {loss} rupees.")
```

```python
def loss_profit_calculate(cost,sell):
    profit = sell - cost
    loss =  cost - sell
    if sell > cost:
        print(f"You are in Profit of {profit} rupees.")

    elif sell == cost:
        print("You are neither in Profit nor in Loss")
    else:
        print(f"You are in loss of {loss} rupees.")

loss_profit_calculate(22,23)
```

## **14.Given a string s representing time in 24-hour format "HH:MM", compute the smallest angle in degrees between the hour and minute hands of an analog clock.**

```python
def clock_angle(time_str):
    # Split the time string into hours and minutes
    hours, minutes = map(int, time_str.split(':'))

    # Calculate the angles
    minute_angle = minutes * 6  # Each minute is 6 degrees
    hour_angle = (hours % 12) * 30 + (minutes / 60) * 30  # Each hour is 30 degrees, plus minute contribution

    # Calculate the absolute difference
    angle = abs(hour_angle - minute_angle)

    # Find the minimum angle
    minimum_angle = min(angle, 360 - angle)

    return minimum_angle

# Example usage
time_str = "14:30"  # Example input
angle = clock_angle(time_str)
print(f"The minimum angle between the hour and minute hands at {time_str} is {angle} degrees.")
```

## **15. Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).**

```python
def rectangles_overlap(L1, R1, L2, R2):
    # Unpack the coordinates
    x1, y1 = L1  # Top-left of rectangle 1
    x2, y2 = R1  # Bottom-right of rectangle 1
    x3, y3 = L2  # Top-left of rectangle 2
    x4, y4 = R2  # Bottom-right of rectangle 2

    # Check for no overlap
    if (x2 < x3 or  # Rectangle 1 is to the left of Rectangle 2
        x1 > x4 or  # Rectangle 1 is to the right of Rectangle 2
        y2 < y3 or  # Rectangle 1 is above Rectangle 2
        y1 > y4):   # Rectangle 1 is below Rectangle 2
        return False  # Rectangles do not overlap

    return True  # Rectangles overlap

# Example usage
L1 = (1, 3)  # Rectangle 1 top-left
R1 = (4, 1)  # Rectangle 1 bottom-right
L2 = (2, 2)  # Rectangle 2 top-left
R2 = (5, 0)  # Rectangle 2 bottom-right

if rectangles_overlap(L1, R1, L2, R2):
    print("The rectangles overlap.")
else:
    print("The rectangles do not overlap.")
```

## **18. Write a program that will check whether the number is armstrong number or not.**


```python
num = int(input("Enter a number: "))

# Convert the number to a string and then create a list of its digits
digits = [int(digit) for digit in str(num)]
n  = len(digits)
results = []
for i in digits:
    power = i**n
    results.append(power)
if sum(results) == num:
    print('The Number is armstrong')
else:
    print('The Number is not armstrong')
```

## **20. Salary Calculation Program**
Write a program that calculates the in-hand salary after applying the following deductions:

**Deductions:**

- HRA: 10%
- DA: 5%
- PF: 3%
  
**Tax Slabs:**

- If the salary is between 5 - 10 lakh: 10%
- If the salary is between 11 - 20 lakh: 20%
- If the salary exceeds 20 lakh: 30%
- If the salary is up to 1 lakh: print "k".

```python
salary = 800000

# Deductions
hra = salary / 100 * 10    # 10% HRA
da = salary / 100 * 5      # 5% DA
pf = salary / 100 * 3      # 3% PF

# Total Deduction
deductions = hra + da + pf

# Remaining Salary after deductions
rem_salary = salary - deductions

# Initializing in-hand salary
in_hand_salary = rem_salary 

# Tax Calculation
if salary > 100000:
    if salary >= 500000 and salary <= 1000000:
        in_hand_salary -= rem_salary / 100 * 10  # 10% Tax
    elif salary >= 1100000 and salary <= 2000000:
        in_hand_salary -= rem_salary / 100 * 20  # 20% Tax
    elif salary > 2000000:
        in_hand_salary -= rem_salary / 100 * 30  # 30% Tax
else:
    print('k')

# Output
print('Final Take-Home Salary: ', in_hand_salary)
```

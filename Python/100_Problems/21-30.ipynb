## **21. Menu-Driven Conversion Program**
**Create a menu-driven program with the following options:**

- Convert centimeters (cm) to feet (ft)
- Convert kilometers (km) to miles
- Convert US Dollars (USD) to PKR
- Exit the program

**Program Flow:**
- Display the menu options to the user.
- Prompt the user to choose an option.
- Based on the user's choice, perform the corresponding conversion:
- For option 1: Implement the conversion from centimeters to feet.
- For option 2: Implement the conversion from kilometers to miles.
- For option 3: Implement the conversion from USD to INR.
- Allow the user to continue making conversions until they choose to exit the program.
- If the user selects option 4, terminate the program.

```python
# Function to convert centimeters to feet
def cm_to_ft(cm):
    return cm * 0.0328084

# Function to convert kilometers to miles
def km_to_miles(km):
    return km * 0.621371

# Function to convert USD to PKR
def usd_to_pkr(usd):
    return usd * 280.1452

while True:
    user_input = input('''
                   Enter 1 for centimeter to feet conversion.
                   Enter 2 for kilometers to miles conversion.
                   Enter 3 for USD to PKR conversion.
                   Enter 4 to exit the program.
                   ''')

    if user_input == '1':
        cm = float(input("Enter the number of centimeters to convert into feet: "))
        result = cm_to_ft(cm)
        print(f"{cm} cm is equal to {result:.2f} ft.")

    elif user_input == '2':
        km = float(input("Enter the number of kilometers to convert into miles: "))
        result = km_to_miles(km)
        print(f"{km} km is equal to {result:.2f} miles.")

    elif user_input == '3':
        usd = float(input("Enter the amount in USD to convert into PKR: "))
        result = usd_to_pkr(usd)
        print(f"{usd} USD is equal to {result:.2f} PKR.")

    elif user_input == '4':
        print("Exiting the program.")
        break 

    else:
        print("Enter a valid input, try between 1 to 4.")
```

### **22.Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.**

```python
heads = int(input("Enter total number of heads: "))
legs = int(input("Enter total number of legs: "))

# Calculate number of dogs and chickens
dogs = (legs - 2 * heads) // 2
chickens = heads - dogs

# Validation
if dogs < 0 or chickens < 0 or (2 * chickens + 4 * dogs) != legs:
    print("Invalid input! No possible solution.")
else:
    print("Number of chickens:", chickens)
    print("Number of dogs:", dogs)
```

## **24. Write a program to find the sum of first n numbers, where n will be provided by the user. Eg if the user provides `n=10` the output should be `55`.**

```python
# Get input from the user
n = int(input("Enter the number: "))
L = []
for i in range(n + 1):
    L.append(i)
total_sum = sum(L)
print(f"The sum of the first {n} numbers is: {total_sum}")
```

## **25.Write a program that can multiply 2 numbers provided by the user without using the `*` operator**

```python
def multiply(a, b):
    c = []
    for i in range(b):
        c.append(a)
    return sum(c)

# User input
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate and display the product
result = multiply(num1, num2)
print(f"The product of {num1} and {num2} is: {result}")
```

## **26.Write a program that can find the factorial of a given number provided by the user.**

```python
def factorial(num):
    if num==0:
        return 1
    else:
        f = 1
        for i in range(1,num+1):
            f = i*f
        print(f)

factorial(4)
```

## **28.Write a program to print whether a given number is prime number or not**

```python
def Prime(num):
    if num <= 1:
        print("Not a Prime Number")
    else:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                print("Not a Prime Number")
                break
        else:
            print(f"{num} is a Prime Number")

Prime(25)
```

## **29.Print all the armstrong numbers in the range of 100 to 1000**

```python
for i in range(100,1001):
    digits = [int(digit) for digit in str(i)]
    n  = len(digits)
    results = []
    for j in digits:
        power = j**n
        results.append(power)
    if sum(results) == i:
        print(i)
```

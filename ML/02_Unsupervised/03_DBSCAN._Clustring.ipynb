## **DBSCAN Clustring**
Density Based Spatial Clustring with Application and Noise

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('./data/shopping-data.csv')
df.head()
```

```python
df = df.iloc[:,3:].values
df
```

```python
from sklearn.cluster import DBSCAN
cluster = DBSCAN(eps=9,min_samples=3)
cluster.fit(df)
cluster.labels_
```

```python
plt.figure(figsize=[10,6])
plt.scatter(df[:,0],df[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
import gensim
import os


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []

# Iterate through each file in the specified directory
for filename in os.listdir("./data/game_of_thrones"):
    # Open the file using the 'with' statement to ensure proper resource management
    with open(os.path.join("./data/game_of_thrones", filename), encoding='latin-1') as f:
        corpus = f.read()  # Read the contents of the file
        raw_sent = sent_tokenize(corpus)  # Sentence tokenization
        
        for sent in raw_sent:
            story.append(simple_preprocess(sent))  # Basic pre-processing of each tokenized sentence
```

```python
from nltk.corpus import stopwords

def remove_stop_words(text):
    # Get the list of English stop words
    stop_words = set(stopwords.words('english'))
    
    # If text is a list, filter out stop words directly
    if isinstance(text, list):
        new_text = [word for word in text if word.lower() not in stop_words]
    else:
        # If text is a string, split it first
        new_text = [word for word in text.split() if word.lower() not in stop_words]
    
    return new_text if isinstance(text, list) else ' '.join(new_text)
```

```python
model = gensim.models.Word2Vec(
    sentences=None,
    window=10,
    min_count=2,
    workers=4
)
```

```python
# Vocabulary
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

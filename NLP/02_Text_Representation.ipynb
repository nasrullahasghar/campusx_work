# **Text Representation**
Text representation in Natural Language Processing (NLP) is the process of converting raw text into numerical formats that computers and machine learning algorithms can process. This is critical because models operate on mathematical vectors rather than characters or words.

- It is also called as `Text Vectorization`/`Feature Extraction`

#  **1. One-Hot Encoding (OHE)**

### 🔹 What is One-Hot Encoding?
One-Hot Encoding (OHE) is a **text representation technique** that converts **categorical text data** into **numerical vectors** so that machine learning models can understand it.

Each **unique word/token** is represented as a vector of **0s and 1s**:
- `1` → token is present  
- `0` → token is absent  

📌 **Important**: OHE does **not** capture meaning or context, it only captures **presence or absence**.

---

### 🔹 How OHE Converts Text into Vectors

Assume we have the following sentences:

```text
"I love AI"
"I love ML"


### **Step 1: Create Vocabulary**

````python
Vocabulary = ["I", "love", "AI", "ML"]
````

### **Step 2: Assign Index to Each Word**
| Word | Index |
| ---- | ----- |
| I    | 0     |
| love | 1     |
| AI   | 2     |
| ML   | 3     |


### **Step 3: Convert Sentences into Vectors**
| Sentence  | Vector       |
| --------- | ------------ |
| I love AI | [1, 1, 1, 0] |
| I love ML | [1, 1, 0, 1] |

    Each sentence is now a binary vector.

### **One-Hot Encoding Using Scikit Learn**

````python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample text data
data = [["I"], ["love"], ["AI"], ["love"], ["ML"]]

# Initialize encoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform
encoded = encoder.fit_transform(data)

# Output
print("Vocabulary:", encoder.categories_)
print("Encoded Vectors:\n", encoded)
````

### **One-Hot Encoding Using CountVectorizer**

````python
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
    "I love AI",
    "I love ML"
]

vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Vectors:\n", X.toarray())
````

### **✅ Advantages & ❌ Disadvantages of OHE**
| Advantages                        | Disadvantages                         |
| --------------------------------- | ------------------------------------- |
| Simple and easy to understand     | Creates very high-dimensional vectors |
| No mathematical assumptions       | No semantic meaning captured          |
| Works well for small vocabularies | Memory inefficient                    |
| Prevents ordinal bias             | Not suitable for large text corpora   |
| Model-friendly for basic ML       | Cannot handle unseen words well       |


📌 Best Used When:

- Vocabulary is small

- Dataset is not very large

- Baseline or educational models

📌 Avoid When:

- Large text data

- Semantic understanding is required

# **2. Bag of Words (BoW)**

### 🔹 **What is Bag of Words?**
Bag of Words (BoW) is a **text representation technique** that converts text into **numerical vectors** based on **word frequency**.

📌 BoW only cares about:
- **Which words appear**
- **How many times they appear**

🚫 It **ignores grammar, word order, and context**.

---

### 🔹 **How BoW Converts Text into Vectors (Simple Explanation)**
Consider these sentences:

```text
"I love AI"
"I love ML"


### **Step 1: Create Vocabulary**

````python
Vocabulary = ["I", "love", "AI", "ML"]
````

### **Step 2: Count Word Occurrences**
| Sentence  | I | love | AI | ML |
| --------- | - | ---- | -- | -- |
| I love AI | 1 | 1    | 1  | 0  |
| I love ML | 1 | 1    | 0  | 1  |

    Each sentence is represented as a count-based vector.

````python
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
    "I love AI",
    "I love ML"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Vectors:\n", X.toarray())
````

# **3. BoW with N-grams (Unigrams + Bigrams)**

````python
# Bigram
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
    "I love AI",
    "I love ML"
]

vectorizer = CountVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Vectors:\n", X.toarray())
````

````python
vectorizer.vocabulary_
````

````python
# Trigram
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
    "I love AI",
    "I love ML"
]

vectorizer = CountVectorizer(ngram_range=(1, 3))
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Vectors:\n", X.toarray())
````

````python
vectorizer.vocabulary_
````

## **BoW vs One-Hot Encoding**
| Feature                | OHE  | BoW        |
| ---------------------- | ---- | ---------- |
| Uses frequency         | ❌ No | ✅ Yes      |
| Vector sparsity        | High | High       |
| Captures importance    | ❌ No | ⚠️ Partial |
| Handles repeated words | ❌ No | ✅ Yes      |

--------------------------------------------------------

## **✅ Advantages & ❌ Disadvantages of BoW**
| Advantages                         | Disadvantages                   |
| ---------------------------------- | ------------------------------- |
| Simple and fast                    | Ignores word order              |
| Captures word frequency            | No semantic meaning             |
| Easy to implement                  | High-dimensional sparse vectors |
| Works well for text classification | Sensitive to vocabulary size    |
| Can use n-grams                    | Fails on unseen words           |


📌 Best Used When:

- Text classification tasks

- Small to medium datasets

- Baseline ML models (Naive Bayes, Logistic Regression)

📌 Avoid When:

- Context and meaning are important

- Large-scale NLP tasks

# **4. TF-IDF (Term Frequency – Inverse Document Frequency)**

## 🔹 **What is TF-IDF?**
TF-IDF is a **statistical text representation technique** that measures how **important a word is to a document** relative to a collection of documents (corpus).

📌 Unlike BoW, TF-IDF:
- Reduces the importance of **common words**
- Increases the importance of **rare but meaningful words**

---

## 🔹 **How TF-IDF Works**

TF-IDF is a product of **two components**:

### **1️⃣ Term Frequency (TF)**
Measures how often a word appears in a document.

```text
TF(word) = (Number of times word appears in a document) 
           --------------------------------------------
           (Total number of words in the document)


### 2️⃣ Inverse Document Frequency (IDF)

Measures how unique a word is across all documents.

IDF(word) = log(`Total number of documents` / `Number of documents containing the word`)



### **Final Formula**

TF-IDF(word) = TF(word) × IDF(word)

### Example (Intuition)

Documents:

    D1: "I love AI"
    D2: "I love ML"

- Word "love" → appears in every document → low IDF

- Word "AI" → appears in only one document → high IDF

📌 TF-IDF assigns higher weight to "AI" and "ML" than to "love".

### **TF-IDF using TfidfVectorizer**

````python
from sklearn.feature_extraction.text import TfidfVectorizer

sentences = [
    "I love AI",
    "I love ML"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("TF-IDF Vectors:\n", X.toarray())
````

### **TF-IDF with N-grams**

````python
from sklearn.feature_extraction.text import TfidfVectorizer

sentences = [
    "I love AI",
    "I love ML"
]

vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("TF-IDF Vectors:\n", X.toarray())
````

### **BoW vs TF-IDF**
| Feature                | BoW   | TF-IDF   |
| ---------------------- | ----- | -------- |
| Uses frequency         | ✅ Yes | ✅ Yes    |
| Penalizes common words | ❌ No  | ✅ Yes    |
| Captures importance    | ❌ No  | ✅ Better |
| Semantic meaning       | ❌ No  | ❌ No     |

### **✅ Advantages & ❌ Disadvantages of TF-IDF**
| Advantages                         | Disadvantages                   |
| ---------------------------------- | ------------------------------- |
| Highlights important words         | Still ignores word order        |
| Reduces effect of stopwords        | No semantic understanding       |
| Works well for text classification | High-dimensional sparse vectors |
| Better than BoW                    | Struggles with synonyms         |
| Fast and scalable                  | Not suitable for deep learning  |

📌 Best Used When:

- Text classification

- Search engines

- Keyword extraction

- Medium-sized NLP problems

📌 Avoid When:

- Context and word relationships matter

- Deep semantic understanding is required

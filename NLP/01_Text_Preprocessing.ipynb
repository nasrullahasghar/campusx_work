## **3. Remove URLs**

```python
# Sample variables with text strings and dummy URLs
text1 = "Explore the beauty of nature. Visit us at http://www.example.com/nature"
text2 = "Discover advancements in technology at http://www.example.com/technology"
text3 = "Join a culinary adventure at http://www.example.com/culinary"
text4 = "Dive into art and culture at http://www.example.com/art"
text5 = "Stay updated with the latest news at http://www.example.com/news"

def remove_urls(text):
    # A simple regex pattern to match http and https URLs
    url_pattern = r'https?://\S+|www\.\S+'
    # Replace all occurrences of the URL pattern with an empty string
    cleaned_text = re.sub(url_pattern, '', text)
    return cleaned_text.strip()  # Remove any leading/trailing whitespace

# Remove URL from text5 and store the cleaned text
cleaned_text5 = remove_urls(text5)

# Print the original and cleaned text
print("Original Text:", text5)
print("Cleaned Text:", cleaned_text5)
```

## **4. Remove Punctuation**

```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
import time
# Very Slow Technique
start = time.time()
text6 = 'string. with. punctuation!'
remove_punctuation(text6)
time1 = time.time() - start
print(time1*50000)
```

```python
# Fast Way
def remove_punc1(text):
    return text.translate(str.maketrans("","",exclude))
```

```python
# Start timing
start = time.time()

# Sample text with punctuation
text6 = 'string. with. punctuation!'

# Call the function to remove punctuation
cleaned_text6 = remove_punc1(text6)

# Calculate the time taken for the operation
time2 = time.time() - start

# Print how long it takes to perform this operation 50,000 times
print("Time taken for 50,000 operations:", time2 * 50000)

# Optionally, print the cleaned text
print("Cleaned Text:", cleaned_text6)
```

```python
df['review'] = df['review'].apply(remove_punc1)
```

## **5. Chat Word Treatment**
- LOL: Laugh Out Loud
- OMG: Oh My God
- BRB: Be Right Back
- BTW: By The Way
- IDK: I Don't Know
- ASAP: As Soon As Possible
- FYI: For Your Information
- TTYL: Talk To You Later
- LMK: Let Me Know
- SMH: Shaking My Head
- TBH: To Be Honest
- JK: Just Kidding
- NVM: Nevermind
- IMO/IMHO: In My Opinion/In My Humble Opinion
- ICYMI: In Case You Missed It
- ROFL: Rolling On the Floor Laughing 

```python
# Dictionary of acronyms and their meanings
chat_words = {
    "LOL": "Laugh Out Loud",
    "OMG": "Oh My God",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "IDK": "I Don't Know",
    "ASAP": "As Soon As Possible",
    "FYI": "For Your Information",
    "TTYL": "Talk To You Later",
    "LMK": "Let Me Know",
    "SMH": "Shaking My Head",
    "TBH": "To Be Honest",
    "JK": "Just Kidding",
    "NVM": "Nevermind",
    "IMO/IMHO": "In My Opinion/In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "ROFL": "Rolling On the Floor Laughing"
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        # Check if the word is in chat_words, ignore case
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])  # Replace with meaning
        else:
            new_text.append(w)  # Keep the original word
    return " ".join(new_text)  # Join the list into a single string

# Example usage
input_text = "LOL that is funny! BRB, I need to grab something."
output_text = chat_conversion(input_text)
print("Original Text:", input_text)
print("Converted Text:", output_text)
```

## **7. Remove the Stop Words**
    "the," "is," "and," "in"

```python
import nltk
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopwords.words('english')
```

```python
def remove_stop_words(text):
    # Get the list of English stop words
    stop_words = set(stopwords.words('english'))
    
    # Split the text into words and filter out stop words
    new_text = [word for word in text.split() if word.lower() not in stop_words]
    
    return ' '.join(new_text)  # Join the list back into a string
```

```python
input_text = "This is an example sentence demonstrating the removal of stop words."
cleaned_text = remove_stop_words(input_text)
print("Original Text:", input_text)
print("Cleaned Text:", cleaned_text)
```

## **8. Handling Emojis**

```python
import re
def remove_emji(text):

    emoji_pattern = re.compile("["
                            u"\U0001F600-\U0001F64F" #emotions
                            u"\U0001F300-\U0001F5FF" #sumbols and pictographs
                            u"\U0001F680-\U0001F6FF" #transport and map symbols
                            u"\U0001F1E0-\U0001F1FF" #flags
                            u"\U00002702-\U000027B0"  
                            u"\U000024C2-\U0001F251" 

                            "]+",flags = re.UNICODE)
    return emoji_pattern.sub(r'',text)
```

```python
input_text = "I love exploring new places! 🌍✨"
cleaned_text = remove_emji(input_text)
print("Original Text:", input_text)
print("Cleaned Text:", cleaned_text)
```

## **10. Stemming**
Stemming in natural language processing (NLP) involves reducing words to their root form or stem, which may not always be a valid word. For example, "arguing" and "argued" may sometimes be stemmed to "argu," which isn't a correct word. This is achieved primarily by removing suffixes.
- NLTK library is used to do this

```python
from nltk.stem.porter import PorterStemmer
```

```python
# Download necessary NLTK resources
nltk.download('punkt')  # For tokenization
```

```python
# Initialize the Porter Stemmer
ps = PorterStemmer()

def stem_words(text):
    # Apply stemming to each word in the text
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
input_text = "walk walks walking walked"
stemmed_text = stem_words(input_text)
print("Original Text:", input_text)
print("Stemmed Text:", stemmed_text)
```

`Here we can see some of the words are not remain English words, this is the main problem of stemming.`

## **11. Lemmatization**

    Lemmatization is a more advanced and accurate form of text normalization compared to stemming. Instead of simply chopping off word endings, lemmatization considers the word’s meaning and part of speech to reduce it to its base or root form (lemma), ensuring the word retains its meaning. For example, it converts “running” to “run” and “better” to “good,” unlike stemming which might incorrectly convert “running” to “runn.”

- Its is Slower then Stemming because it searches for Words don't just cut off the word.
- When we want to show the output to user we will use  Lemmatization.
- When we just need speed we will use Stemming.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
text = "The cats were running faster than the dogs."
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations = ".!?,.:"

sentence_words = nltk.word_tokenize(sentence)
print("{:20} {}".format('Word','Lemma'))
for word in sentence_words:
    if word not in punctuations:
        print('{:20} {}'.format(word, wordnet_lemmatizer.lemmatize(word,pos='v')))
```

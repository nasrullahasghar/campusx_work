```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("I will google about facebook")
```

```python
doc.text
```

```python
doc[-1]
```

```python
doc[2].pos_
```

```python
doc[2].tag_
```

```python
spacy.explain("VB")
```

```python
for word in doc:
    print(word.text,"=======>",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc2 = nlp("I left the room")
for word in doc2:
    print(word.text,"=======>",word.pos_,"=======>",word.tag_,spacy.explain(word.tag_))
```

```python
doc3 = nlp("To the left of the room")
for word in doc3:
    print(word.text,"=======>",word.pos_,"=======>",word.tag_,spacy.explain(word.tag_))
```

```python
doc4 = nlp("I read books on history")
for word in doc4:
    print(word.text,"=======>",word.pos_,"=======>",word.tag_,spacy.explain(word.tag_))
```

```python
doc5 = nlp("I have read books on history")
for word in doc5:
    print(word.text,"=======>",word.pos_,"=======>",word.tag_,spacy.explain(word.tag_))
```

```python
from spacy import displacy
```

```python
doc6 = nlp("The quick brown fox jumps over the lazy dog")
displacy.render(doc6, style='dep', jupyter=True)
```

```python
options={
    'distance':80,
    'compact':True,
    'color':'#fff',
    'bg':'#00a65a'
}
displacy.render(doc6,style='dep',jupyter=True,options=options)
```
